## Ejercicio #1
Código personal que calcule transformada de Fourier de una función f(x).

```python
def generate_periodic_random_array(period_length, num_periods, random_seed=None):
    """
    Generate an array of periodic random numbers.
    
    Parameters:
    - period_length (int): Length of one period.
    - num_periods (int): Number of periods to repeat the random sequence.
    - random_seed (int or None): Seed for the random number generator for reproducibility.
    
    Returns:
    - numpy.ndarray: A periodic array of random numbers.
    """
    if random_seed is not None:
        np.random.seed(random_seed)
    
    # Generate one period of random numbers.
    one_period = np.random.rand(period_length)
    
    # Tile the one-period array to create the periodic pattern.
    periodic_array = np.tile(one_period, num_periods)
    
    return periodic_array

# Parameters
period_length = 10   # Length of one period
num_periods = 5      # Total number of periods
random_seed = 42     # Seed for reproducibility (optional)

# Generate the periodic random array.
array = generate_periodic_random_array(period_length, num_periods, random_seed)

# Display the array.
print("Periodic random array:")
print(array)

# Plot the array to visualize the periodicity.
plt.figure(figsize=(10, 4))
plt.plot(array, marker='o')
plt.title("Periodic Random Array")
plt.xlabel("Index")
plt.ylabel("Random Value")
plt.grid(True)
plt.show()

array
```

```python
def periodic_rand(period_length, num_periods, random_seed=None): #Return an array of periodic random numbers 
    if random_seed is not None:
        np.random.seed(random_seed)
    one_period = np.random.rand(period_length)
    periodic_array = np.tile(one_period, num_periods)
    return periodic_array

def discrete_FT( func , down = -100 ,up = 100 , range_data = 50, DATA = False, data = []):
    def functional(F):
        x = sp.symbols('x') # Variable a evaluar
        expr = sp.sympify(F) # Conversor de string a sympy
        eval = sp.lambdify(x, expr, modules='numpy') # Conversor de simbolico a numerico
        DP(Math(r"\text{Funcion f(x): }\,\,"+ sp.latex(expr))) # Mostrar la ecuacion a evaluar
        return eval
    def X_k(values):
        x_k = []
        N = len(values)
        for k in range(N):
            tmp = 0
            for b in range(N):
                tmp += values[b] * np.exp(-1j * 2 * np.pi * (k / N) * b)
            x_k.append(tmp)
        x_k  = np.array(x_k)
        return x_k

    if DATA == False:
        F_eval = functional(func) # Funcion proveniente de Sympy evaluando numericamente
        data = np.linspace(down,up,range_data) # Rango de datos
        x_n = np.array(F_eval(data))
        txt = 'Funcion original'
    else:
        x_n = np.array(data)
        txt = 'Datos originales'

    plt.figure(figsize=(14, 7))
    plt.subplot(1,2,1)
    plt.plot(data,x_n)
    plt.title(txt)
    
    plt.subplot(1,2,2)
    aliasing = int(range_data/2)
    plt.plot(abs(X_k(x_n)[0:aliasing]))
    plt.yscale('log')
    plt.title("Transformada Discreta de Fourier")
    return 'Done'
    
discrete_FT("30*sin(pi*x+cos(x))",down = -100 ,up = 100 , range_data = 500)
```
